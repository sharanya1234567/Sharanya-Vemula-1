```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/Online Sales Data (1).csv')

# Check for null values
print("Null values in each column:")
print(df.isnull().sum())

# Handle null values
df_cleaned = df.dropna()  # Dropping rows with any null values

# Save the cleaned dataset
df_cleaned.to_csv('online_sales_cleaned.csv', index=False)

product_sales = df_cleaned.groupby('product name')['sales'].sum()

plt.figure(figsize=(10, 6))
product_sales.plot(kind='bar')
plt.title('Total Sales per Product')
plt.xlabel('Product Name')
plt.ylabel('Sales')
plt.show()


sales_over_time = df_cleaned.groupby('date')['sales'].sum()

plt.figure(figsize=(10, 6))
sales_over_time.plot(kind='line')
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df_cleaned['sales'], bins=20, edgecolor='black')
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df_cleaned.boxplot(column='sales', by='product')
plt.title('Box Plot of Sales per Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.suptitle('')
plt.show()
```

```python
product_sales = df_cleaned.groupby('product')['sales'].sum()
plt.figure(figsize=(10, 6))
product_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales Share by Product')
plt.ylabel('')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('online_sales.csv')

# Check for null values
print("Null values in each column:")
print(df.isnull().sum())

# Handle null values
df_cleaned = df.dropna()  # Dropping rows with any null values

# Save the cleaned dataset
df_cleaned.to_csv('online_sales_cleaned.csv', index=False)

# Visualization: Bar chart of sales per product
product_sales = df_cleaned.groupby('product')['sales'].sum()

plt.figure(figsize=(10, 6))
product_sales.plot(kind='bar')
plt.title('Total Sales per Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()

# Visualization: Line chart of sales over time
sales_over_time = df_cleaned.groupby('date')['sales'].sum()

plt.figure(figsize=(10, 6))
sales_over_time.plot(kind='line')
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()

# Additional Visualization: Histogram of Sales
plt.figure(figsize=(10, 6))
plt.hist(df_cleaned['sales'], bins=20, edgecolor='black')
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()

# Additional Visualization: Scatter Plot of Sales vs. Quantity
plt.figure(figsize=(10, 6))
plt.scatter(df_cleaned['quantity'], df_cleaned['sales'])
plt.title('Sales vs. Quantity')
plt.xlabel('Quantity')
plt.ylabel('Sales')
plt.show()

# Additional Visualization: Box Plot of Sales per Product
plt.figure(figsize=(12, 6))
df_cleaned.boxplot(column='sales', by='product')
plt.title('Box Plot of Sales per Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.suptitle('')
plt.show()

# Additional Visualization: Pie Chart of Sales Share by Product
product_sales = df_cleaned.groupby('product')['sales'].sum()
plt.figure(figsize=(10, 6))
product_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales Share by Product')
plt.ylabel('')
plt.show()
```
